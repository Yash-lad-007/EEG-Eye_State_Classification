```python
# 1. Import Required Libraries
# ===============================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

from scipy import stats
```

```python
# 2. Load EEG Dataset
# ===============================
# Example: UCI EEG Eye State dataset
# Replace file name with your dataset
df = pd.read_csv("/content/EEG_Eye_State_Classification.csv")

print("Dataset Shape:", df.shape)
print(df.head())
```

```python
# 3. Separate Features and Labels
# ===============================
X = df.iloc[:, :-1]   # EEG channels
y = df.iloc[:, -1]    # Class label
```

```python
# 4. Handle Missing Values
# ===============================
print("\nMissing values before:")
print(X.isnull().sum())

X.fillna(X.mean(), inplace=True)

print("\nMissing values after:")
print(X.isnull().sum())
```

```python
# 5. Outlier Removal using Z-score
# ===============================
z_scores = np.abs(stats.zscore(X))
X = X[(z_scores < 3).all(axis=1)]
y = y[X.index]

print("\nShape after outlier removal:", X.shape)
```

```python
# 6. Label Encoding (if labels are categorical)
# ===============================
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
# 7. Feature Scaling (IMPORTANT for KNN)
# ===============================
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# 8. Train-Test Split
# ===============================
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
# 9. KNN Model Training
# ===============================
knn = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
knn.fit(X_train, y_train)
```

```python
# 10. Model Prediction
# ===============================
y_pred = knn.predict(X_test)
```

```python
# 11. Model Evaluation
# ===============================
accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
cr = classification_report(y_test, y_pred)

print("\nAccuracy:", accuracy)
print("\nConfusion Matrix:\n", cm)
print("\nClassification Report:\n", cr)
```

```python
# 12. Accuracy vs K Value (Optional)
# ===============================
accuracy_list = []

for k in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    pred = knn.predict(X_test)
    accuracy_list.append(accuracy_score(y_test, pred))

plt.plot(range(1, 21), accuracy_list)
plt.xlabel("K Value")
plt.ylabel("Accuracy")
plt.title("Accuracy vs K Value")
plt.show()
```
